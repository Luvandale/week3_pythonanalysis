```
import pandas as pd
import numpy as np
import csv
```

```
#reading our dataset files
data1=pd.read_csv("Telcom_dataset.csv")
data2=pd.read_csv("Telcom_dataset2.csv")
data3=pd.read_csv("Telcom_dataset3.csv")
```

```
#checking information about our files
print(data1.info())
print("\n")
print(data2.info())
print("\n")
print(data3.info())
```

```
#standardize column name since there are spelling mistakes and other discrepancies in the column names for the first data
data1.rename(columns={"PRODUTC":"PRODUCT","VALUE":"VALUE","DATETIME":"DATE_TIME","CELL_ON_SITE":"CELL_ON_SITE","DW_A_NUMBER_INT":"DW_A_NUMBER","DW_B_NUMBER_INT":"DW_B_NUMBER","COUNTRY_A":"COUNTRY_A","CELL_ID":"CELL_ID","SITE_ID":"SITE_ID"},inplace=True)
data1.columns
```

```
#standardize column name since there are spelling mistakes and other discrepancies in the column names for the second data
data2.rename(columns={"PRODUCT":"PRODUCT","VALUE":"VALUE","DATE_TIME":"DATE_TIME","CELL_ON_SITE":"CELL_ON_SITE","DW_A_NUMBER":"DW_A_NUMBER","DW_B_NUMBER":"DW_B_NUMBER","COUNTRY_A":"COUNTRY_A","CELL_ID":"CELL_ID","SITE_ID":"SITE_ID"},inplace=True)
data2.columns
```

```
#standardize column name since there are spelling mistakes and other discrepancies in the column names for the third data
data3.rename(columns={"PRODUCT":"PRODUCT","VALUE":"VALUE","DATE_TIME":"DATE_TIME","CELL_ON_SITE":"CELL_ON_SITE","DW_A_NUMBER_INT":"DW_A_NUMBER","DW_B_NUMBER_INT":"DW_B_NUMBER","COUNTRY_A":"COUNTRY_A","CELLID":"CELL_ID","SIET_ID":"SITE_ID"},inplace=True)
data3.columns
```

```
#merging our three datasets based on column names
merged_data = pd.concat([data1,data2,data3],ignore_index=True)
```

```
#checking our merged data
merged_data.info()
```

```
# showing the first five rows
merged_data.head()
```

```
# cleaning data(finding if there exists duplicates)
merged_data.duplicated()
```

```
# dropping the duplicates
merged_data.drop_duplicates()
```

```
# checking for null values
merged_data.isnull()
```

```
# figuring out the most used product
most_used_product = merged_data.PRODUCT.mode()
print(most_used_product)
```

```
# figuring out the statistics of specific products
# grouping value by product
groupby_product = merged_data['VALUE'].groupby(merged_data['PRODUCT'])
groupby_product
```

```
# descriptive statistics by group
merged_data['VALUE'].groupby(merged_data['PRODUCT']).describe()
```

```
#  find the sum of the prices of the product 
# shows the total billing price.
groupby_product.sum()
```

```
# shows that sms has a higher billing price based on the few samples that can be seen
merged_data['VALUE'].groupby([merged_data['PRODUCT'], merged_data['DATE_TIME']]).sum()
```

```
#  trying to see the number of voice and sms at different days and times
merged_data.groupby(['PRODUCT', 'DATE_TIME']).count()
```

```
#getting data
data4=pd.read_csv("cells_geo.csv",sep=";")
#show first five rows of our data set
data4.head()
```

```
# cleaning data(finding if there exists duplicates)
data4.duplicated()
```

```
# checking for null values
data4.isnull()
```

```
# dropping null values
drop_null = data4.dropna()
drop_null
```

```
#using scipy to show most used city
from scipy import stats
most_used_city=stats.mode(data4.VILLES)
print(most_used_city[0])
```

```
#using pandas to get the mode of the VILLES column,i.e most comon value in the column
most_used_city = data4.VILLES.mode()
print(most_used_city)
```

```
# checking cities in service
data4.groupby(['VILLES', 'STATUS']).sum()
```

```
# shows the top two cities in services
data4['VILLES'].groupby([data4['LOCALISATION'], data4['STATUS']]).describe()
```

```
# combining our geographical data and telecom data
combined_data =pd.concat([merged_data,data4],ignore_index=True)
```

```
# checking our combined data
combined_data.info()
```

```
# figuring out the cities that were most used during specific hours
cities_mostused = combined_data['VILLES'].groupby(combined_data['DATE_TIME'])
cities_mostused
```
